```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "movies_data.csv"
movie_data_df = pd.read_csv(csv_file)
movie_data_df.head()
```

```python
new_movie_data_df = movie_data_df[['name', 'genre', 'year']].copy()
new_movie_data_df.rename(columns={'name': 'title'}, inplace=True)
new_movie_data_df
```

```python
from pathlib import Path  
filepath = Path('clean_movie_genre_dt.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
new_movie_data_df.to_csv(filepath, index =False)  
```
